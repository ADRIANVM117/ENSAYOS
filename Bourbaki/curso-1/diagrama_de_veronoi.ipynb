```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.spatial import Voronoi, voronoi_plot_2d
```

```python
## Algunos puntos random 

#np.random.seed(12)
puntos = np.random.rand(15,2)


#agregamos puntos falsos 
puntos = np.append(puntos, [[999,999], [-999, 999], [-999, -999], [999, -999]], axis = 0)


vor = Voronoi(puntos)

# Graficamos 
voronoi_plot_2d(vor)

#color 
for region in vor.regions:
    if not -1 in region:
        polygon = [vor.vertices[i] for i in region]
        plt.fill(*zip(*polygon))
        
# Restringir el grafico 

plt.xlim([0,1])
plt.ylim([0,1])


plt.show()
```
