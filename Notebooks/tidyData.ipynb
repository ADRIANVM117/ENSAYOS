## Ejemplo de formato long a formato wider

```python
wider_to_long = pd.read_csv('wider.csv')
wider_to_long 
wider_to_long = wider_to_long.drop(columns = "Unnamed: 0")
wider_to_long
```

```python
#pasamos a formato wider 
wider_to_long.pivot_table(index = ["pais", "anio"], columns = "tipo", values = "cuenta").rename_axist(None, axis = 1)
```
