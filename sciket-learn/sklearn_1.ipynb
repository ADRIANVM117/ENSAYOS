```python
import pandas as pd
df = pd.read_csv('https://assets.datacamp.com/production/repositories/628/datasets/35a8c54b79d559145bbeb5582de7a6169c703136/house-votes-84.csv')
df
```

```python
#Importar KNeighborsClassifier de sklearn.neighbors

from sklearn.neighbors import KNeighborsClassifier

#creamos dos arreglos para las caracteristicas y la variable de respuesta

y = df['republican'].values # <-- values para que los datos pasen a ser un arreglo 
X = df.drop('republican', axis = 1).values

# Crear un knn clasificador con 6 vecinos 

knn = KNeighborsClassifier(n_neighbors =6 )

# Ajustamos el clasificador a los datos 

knn.fit(X,y)
```
