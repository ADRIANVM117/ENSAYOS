```python
import pandas as pd 
datos = pd.read_csv('sales_subset.csv')
print(display(datos))
```

<b> Eliminar duplicados es una habilidad esencial para obtener recuentos mucho mas precisos, ya que la mayor parte del tiempo no se quiere contar la misma variable dos veces </b> 

- Para eliminar valores duplicados utlizamos la función

<b>.drop_duplicates(subset = 'vaiable')  </b>

```python
#Ejemplo 
# Declaramos una variable dondo seleccionamos nuestro DataFrame y aún lado la función 
#e indicamos de que columna queremos que los valores repetidos seán eliminados
eliminar_duplicados = datos.drop_duplicates(subset = 'store')
print(display(eliminar_duplicados.head()))
```

```python
# Podemos indicar en mas de dos columnas los 
# los valores repetidos seán eliminados
eliminar_duplicados = datos.drop_duplicates(subset = ['store', 'type'])
print(display(eliminar_duplicados.head()))
```

- Tambien podemos hacer subconjuntos que cumpplan condiciones


```python
# Filtramos de la columna 'is_holiday' los valores False y a todos esos, eliminamos los datos 
#repetidos en la columna 'date'
eliminar =datos[datos['is_holiday' ] == False ].drop_duplicates('date')
print(display(eliminar.head()))
```

- Contar es una buena manera de tener una visión general de su DataFrame y poder ver peculiaridades en este
- Para contar valores repetidos utilizamos  
<b> .value_counts() </b>

```python
# Hacemos un conteo de los valores repetidos en la colummna 'is_holiday'
conteo_general = datos['is_holiday'].value_counts()
print(conteo_general)
```

<b> Tambien podemos contar la proporción de datos utilizando normalize = True </b>

```python
conteo_proporcion = datos['is_holiday'].value_counts( normalize = True )
print(conteo_proporcion)
```

<b> Podemos ordenar utilizando ascending = True (contar de forma ascendente) o  ascending = False  (contar de forma descendente) </b>

```python
# Contamos el numero de valores repetidos en la columna 'department' y los colocamos de forma
# ascendente 
conteo_ordenado = datos['department'].value_counts( ascending = True )
print(conteo_ordenado)
```

<b> ordenando y sacando la proporción a cada variable repetida   </b>

```python
proporcion_ordenado = datos['department'].value_counts( sort = True, normalize = True )
print(proporcion_ordenado)
```
