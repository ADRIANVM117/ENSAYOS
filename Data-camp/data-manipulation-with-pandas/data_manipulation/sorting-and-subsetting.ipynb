```python
import pandas as pd
import numpy as np

datos = pd.read_csv('data-text.csv')
print(display(datos.head()))
```

> Ordenas columnas

> Para usarlo tienes que tener ya definido tu Dataframe

<b> .sort_values('columna') </b>

```python

datos_orde = datos.sort_values('Numeric')  # Pedimos que la columna 'Numeric' sea ordenada 
print(display(datos_orde.head()))                 #Imprimimos el DataFrame legible 
```

 > #### Para obtener varias columnas ordenadas, dependiendo criterio

```python
datos1 = datos.sort_values(['Year', 'Numeric'])   # Ordenando dos filas 'Year' y 'Numeric'
print(display(datos1.head())) 
```

 > Para ordenar de fomar ascendente o descendente
 
 <b> .sort_values(['columna'], ascending = True o False) <b/>

```python
datos_ordenados = datos.sort_values('Year', ascending=True) #seleccionamos la columna 'year' y 
                                                             # Orden ascendente 
print(display(datos_ordenados))
```

### Tambien podemos hacerlo con varias columnas 


```python
# Selecionamos las columnas 'Year' y 'Country'  y que sean descendente y ascendente en ese orden   
# como seleccionamos True para 'Country' esta se ordena de forma alfabetica 
datos_ordenados2 = datos.sort_values(['Year', 'Country'], ascending = [False, True])
print(display(datos_ordenados2.head()))
```

> Seleccionar Columnas especificas 

<b> Datos[ ['columna'] ]</b>

```python
columna_Year = datos[['Year']]
print(display(columna_Year.head()))
```

### Podemos seleccionar varias columnas 


```python
columna_Year_y_Sex = datos[['Year', 'Sex']]
print(display(columna_Year_y_Sex.head()))
```

- Podemos filtrar datos de las columnas 

```python
filtrar_male = datos[datos['Sex'] == 'Male']  # Filtrar solo los que cumplan con ser 'male' en 
                                              # la columna 'Sex'
print(display(filtrar_male.head()))
```

- Tambien filtrar con varias condiciones  

```python
 # Filtrar solo los que cumplan con ser 'male' en  la columna 'Sex' y 'Australia' de la columna
 # 'Country'  

filtrar_male_y_Australia = datos[(datos['Sex'] == 'Male') & (datos['Country'] == 'Australia')] 
print(display(filtrar_male_y_Australia.head()))
```

- Filtrar por condición >, > = 



```python
 # Filtrar solo los que cumplan con 'Display Value' > 50

filtrar_Display = datos[datos['Display Value'] > 50  ] 
print(display(filtrar_Display.head()))
```

### Otro ejemplo 

```python
varias_condiciones = datos[(datos['Display Value'] > 50) & (datos['Sex'] == 'Female')]

print(display(varias_condiciones.head()))
```

- Subconjunto por variables catagoricas 

> Para  pedirle a una columna que nos regrese mas de una condición utilizamos 

### <b> .isin()  </b>

```python
años = [2000, 1900, 2012]  # Indicamos que condiciones queremos que cumpla 

# Filtramos

años_filtrar = datos[datos['Year'].isin(años)]

print(display(años_filtrar.head()))

#Nos imprime solo los renglones que cumplan con el valor 'Year' que se encuentre en años 
```
