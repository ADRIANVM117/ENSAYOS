- Mutar o transformar un DataFrame basicamente es agregar columnas al mismo DataFrame

```python
import pandas as pd 

datos = pd.read_csv('data-text.csv')
print(display(datos.head()))
```

<b> Añadir columnas </b>

```python
# Agregamos una nueva columna utilizando ya una del DataFrame y recordemos que al final un 
# DataFrame funciona como un arreglo en donde podemos hacer operaciones con cada uno de sus 
#renglos, entonces multiplicamos por 10 y obtenemos una nueva columna 
datos['Numeric * 10'] = datos['Numeric'] * 10

print(display(datos.head()))
```
