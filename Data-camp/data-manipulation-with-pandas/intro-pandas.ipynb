### <b> Clasificando y filtrando datos </b>

> Podemos ordenar los columnas utilizando el metodo .sort_values()

```python
datos.sort_values('Year') #Nos ordena el DF del año mas viejo al mas reciente
```

### <b> Para imprimir los 5 primeros renglones del DF se usa .head()</b>

```python
años = datos.sort_values('Year')  # Filtramos el requisito que queramos 
años.head()                       # Sacamos los 5 primeros que queramos
print(display(años.head()))       # los ponemos en un formato legible usando display
```
