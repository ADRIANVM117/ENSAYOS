- La segmentación del indice. Permite seleccionar elementos consecutivos de un objeto utilizando la sintaxis primero:ultimo. 
- Un DataFrame puede ser cortado por valores dentro del indice o por numero de fila/columna

### <b> Metodo .loc[] </b>
> Es necesario tener esto en cuenta al momento de usar .loc[]

<ol>
    <li> Solo se puede segmentar o cortar un indice si este esta ordenado (usando .sort_index() )</li>
    <li> Para corta en el nivel externo 'primero':'ultimo' puede ser cadenas</li>
    <li> Para cortar en niveles internos ''primero'ultimo' DEBEN ser tuplas</li>
    <li> Si se pasa un unico corte a .loc[] se cortaran en filas</li>
</ol>

```python
import pandas as pd 
datos = pd.read_csv('temperatures.csv')
print(display(datos.head()))
```

- Ordenar indice utilizando .sort_index()

```python
#Fijamos las columnas 'country' y 'city' como nuestros indices externo e interno
datos1 = datos.set_index(['country', 'city'])
#Ordenamos nuestro DataFrame
datos_ordenados = datos1.sort_index()
print(display(datos_ordenados))
```

```python
# Ejemplo 1
#cortar desde 'Pakistan' a 'Russia'
print(display(datos_ordenados.loc['Pakistan':'Russia']))
```

```python
# Ejemplo 2
#cortar desde 'Lahore' a 'Moscow'
print(display(datos_ordenados.loc['Lahore':'Moscow']))
```

### <b> OBSERVACIÓN </b>
- En un indice interno al momento de cortar, te dara el segmento deseado, para cortar un indice interno se utilizan tumplas en donde se selecciona indice externo, indice interno


```python
# Para hacer de 'lahore' a 'Moscow' seria: 
print(display(datos_ordenados.loc[('Pakistan','Lahore'):('Russia','Moscow')]))
```

#### Para seleccionar todos los renglones y ciertas columnas utilizamos 

 <b> [: , 'columna1':'columna 2' ] </b>

```python
#Seleccionamos todaos los renglones  y solo dos columnas 'date' y 'avg_temp_c' 

print(display(datos_ordenados.loc[:,'date':'avg_temp_c']))
```

#### Para hacer un corte especifico de renglones con columnas especificas utilizamos:

<b> [(inicio_indice): (final_indice) ,'columna 1' : 'columna 2' ]</b>

```python
#Hacemos un corte desde 'Pakistan','Lahore' hasta 'Russia','Moscow' selecionando solo las 
#columnas 'date' y 'avg_temp_c'

print(display(datos_ordenados.loc[('Pakistan', 'Lahore'):('Russia', 'Moscow'),
                                  'date': 'avg_temp_c']))
```

#### Ajustar una columna como un indice y ordenarlo 

```python
date_indice_y_ordenados = datos.set_index('date').sort_index()
print(display(date_indice_y_ordenados))
```
