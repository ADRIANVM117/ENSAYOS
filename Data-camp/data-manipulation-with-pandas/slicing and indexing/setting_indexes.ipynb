```python
import pandas as pd 
datos = pd.read_csv('temperatures.csv')
print(display(datos.head()))
```

<b> Pandas </b> te permite asignar tus propias columnas como indice, ayudando a hacer mas organico el procedimiento a la hora de extraer subconjuntos 

- Para volver una columna del DataFrame el indice del mismo. Se utiliza 
#### <b> .set_index('nombre de la columna') <b/>

```python
# Seleccionamos la columna 'date' como el nuevo indice del DataFrame 
datos_indice_nuevo = datos.set_index('date')
print(display(datos_indice_nuevo.head()))
```

- Regresar a su indice original con el metodo 
 #### <b>.reset_index()  </b> 

```python
print(display(datos_indice_nuevo.reset_index()))
```

- Para eliminar la columna utilizada como indice 
#### <b> .set_index(drop = True) </b>

```python
print(display(datos_indice_nuevo.reset_index(drop = True)))
```

#### Un metodo de subconjuntos es .loc[] que acepta valores de indice, cuando se le pasa un unico argumento, tomara un subconjunto de filas 

- Para hacer un subconjunto simple fltrando cierto valores se utiliza 
#### <b> .isin() </b>

```python
# filtramos por medio de la tabala 'city' solo los valores que queremos. En este caso devuelve
# los valores 'Moscow' y 'Paris'
datos1 =  datos[datos['city'].isin(['Moscow','Paris'])]
print(display(datos1))
```

- Tambien se puede hacer una lista de que valores se quieren en el subconjunto y agregarla al metodo: 
#### <b>.isin()  </b>

```python
#Seleccionamos los valores que queremos filtrar
ciudades = [ 'Paris', 'Xian']

print(display(datos[datos['city'].isin(ciudades)]))
```

```python
# Otro ejemplo
# Volvemos la columna 'city' en el indice para buscar con la lista de valores ya hecha sin usar
#el metodo .isin()
indice_pais = datos.set_index('city')
print(display(indice_pais.loc[ciudades]))
```

- Fijar distintos indices 



```python
#fijamos dos  columnas como nuestros indices 
varios = datos.set_index(['country', 'city'])

#declaramos una tupla para hacer el subconjunto
filtrar = [('Brazil', 'Rio De Janeiro'),('Pakistan', 'Lahore')]

print(display(varios.loc[filtrar]))
```

#### Ordenando de renglones por el indice

- Ademas de existir <b> .sort:values() </b> para ordenar por valores,  tambien podemos usar <b> .sort_index() </b> para ordenar por indice 

```python
#utilizamos el DataFrame con el indice de varias columnas ya hecho en el anterior ejemplo 
# y usamos el metodo .sort_index() para ordenar de forma alfabetica el DataFrame

print(display(varios.sort_index()))
```

- NOTA : podemos ver que nos ordena por alfabeto, pero solo el indice externo, para ordenar por el indice interno usamos level = 'indice' para especificar porque indice queremos que se ordene

```python
# ordenar por indice interno de forma alfabetica
print(display(varios.sort_index(level = ['city'])))
```

- Para ordenar por los dos indices utilizamos ascending 

```python
#ordena por indice externo de forma descendente empezando por la ultima letra del alfabeto 
print(display(varios.sort_index(level = ['country', 'city'], ascending = [False, True])))
```
