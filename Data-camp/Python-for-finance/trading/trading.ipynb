<b> un trader puede vender un valor (prestado) a un mayor precio y volver a comprarlo a un precio mas bajo, conocido como posicionarte en CORTO </b>

<b> El traiding </b> se centra en tendencias del mercado a corto plazo y trata de beneficiarse de la volatilidad y las fluctuaciones de los precios. 

> Generalmente un trader analiza los patrones historicos de los datos comerciales e implementa el comercio estrategias basadas en indicadores, señales y reglas para beneficiarse de posibles movimientos futuros de los precios

> Para desempeñarse como trader principalmente trabajara con datos de series temporales de precios de valores

<b> series de tiempo </b> : los datos de series de tiempo son una secuencia de puntos de datos indexados en orden de tiempo


<b> import plotly.graph_objects as go </b> : trazar graficos de velas con el paquete de python 

```python
import pandas as pd 
import numpy as np
#Cargamos datos 

apple_datos = pd.read_csv('apple_datos.csv', index_col = 'Fecha', parse_dates = True)
print(apple_datos)
```

<b> * Ordenamos el DataFrame con la función .sort_values() </b>

```python
pd_apple = apple_datos.sort_values(["Fecha"])
print(pd_apple)
```

###   <b> Graficamos el precio de cierre  </b>


```python
import matplotlib.pyplot as plt 
plt.plot(pd_apple['Cierre'], color = 'red', )
plt.title('precio de cierre DIARIO')
plt.show()
```

> Podemos calcular el cambio porcentual del precio utilizando .pct_change()*100

```python
pd_apple['cambio porcentual'] = pd_apple['Cierre'].pct_change()*100
print(pd_apple)
```
