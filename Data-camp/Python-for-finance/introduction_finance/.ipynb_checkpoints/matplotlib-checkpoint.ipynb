#### Matplotlib es una libreria utilizada para representar datos de manera visual en graficas y sean entendibles para el publico en general.

```python
import matplotlib.pyplot as plt #importamos matplotlib  como plt
# Funciona con listas y tambien arreglos
days = [ 'lun', 'mar', 'mier', 'juev', 'vier', 'sab' ]       
ganancias = [ 232.34, 123.32, 434.34, 654.74, 765.28, 987.43 ]

# representamos la relacion dia-precio en una grafica  de color azul y estilo de linea "--"
plt.plot(days, ganancias, color = 'blue', linestyle = '--')  # Si queremos que la linea sea 
                                                             # solida no agregamos linestyle
plt.show()  #imprimos la grafica
```

 Tipos de linea:  | Escritura
:--------: | -------
Solida     | '-'
pausada    | '--'
pausada con punto | '-.'
punteado   | ':'

#### Añadir etiquetas a una grafica 

```python
plt.plot(days, ganancias, color = 'red', linestyle = ':')

#Añadir etiquetas 
plt.xlabel("Dias")  #Etiqueta para rango x
plt.ylabel("ganancias") #Etiqueta para rango y
plt.title("Ganancias por dia de la semana:") # Titulo de la grafica

plt.show()
```

#### Hacer dos graficas dentro de una misma 

```python
# Podemos comparar dos graficas, declarando dos 
#En este ejemplo graficaremos las ganancias de dos semanas diferentes 
# Utilizamos las listas del ejemplo pasado
ganancias2 = [ 324.34, 178.32, 500.34, 300.74, 532.28, 175.43 ] #Declaramos las ganancias de la semana 2
plt.plot(days, ganancias, color = 'red', linestyle = ':')
plt.plot(days, ganancias2, color = 'blue', linestyle = ':')
#Añadir etiquetas 
plt.xlabel("Dias")  #Etiqueta para rango x
plt.ylabel("ganancias") #Etiqueta para rango y
plt.title("Ganancias por dia de la semana:") # Titulo de la grafica

plt.show()
```

```python
#Otra forma de representar la grafica 
# Scatterplots

plt.scatter(x = days, y = ganacias, color = 'green')
plt.show()
```
