>  Llamadas tambien funciones anonimas 
   - Una función anónima, como su nombre indica es una función sin nombre
   - sirven para realizar tareas sencillas e improvisadas 

>  Mas o menos asi seria su cuerpo general de una función anonima o lambda: >
   
   - Se declara lambda primero, despues las variables a ocupar y por ultimo la tarea a ejecutar
   
     <b> funcion = ( lambda x, y : x + y)  </b>

```python
# Ejemplo  

funcion = (lambda x,y : x + y)

funcion(2,1)
```

```python
# EJEMPLO 2 

echo_word = (lambda palabra, echo: palabra* echo)

repetido = echo_word('\nhola', 3)
print(repetido)
```

#### NOTA: 
- Uno de los mejores usos que se le puede dar a las funciones lambda, es cuando quieres que estas funciones sean incrustradas anonimamente dentro de expresiones mas grandes. Esto significa que la funcionalidad no se almacena en el entorno, a diferencia de una función definida con def. 


#### map()
- <b> map () </b> aplica una función sobre un objeto como una lista.

```python
# EJEMPLO 3w
lista = [ 'accion', 'bono', 'compra', 'valor']

fun = map(lambda item : item + '!!!', lista)

fun_lista = list(fun)

print(fun_lista)
```

#### filter()
-  filter() es una funcion para filtrar elementos que no satisfacen un creiterio  

<b> filter( función, lista ) <b>

```python
#EJEMPLO

# lista de cadena de caracteres  
LISTA = ['zorro', 'carro', 'azul', 'doctora', 'chango', 'mono', 'despues']

#usamos la funcion filteraplicada a una función lambda:
# lo que hace sera contar el numero de elementos en cada cadena dentro de la lista
# y aquellos que cumplan con la condición > 6 caracteres son regresados
funcion_filter = filter(lambda filtro: len(filtro) > 6, LISTA)

# guardamos en formato lista 
result = list(funcion_filter)

print(result)
```

#### reduce()

- Esta función se utiliza para calculos acumulativos y devuelve un unico valor como resultado
- Para usar reduce(), debe importase al modulo functools


```python
from functools import reduce

lista2 = [ 'yo', 'tu', 'el', 'ellos']

result2 = reduce(lambda item1, item2: item1 + item2, lista2)

print(result2)
```

```python
# Ejemplo 
import pandas as pd 
datos = pd.read_csv('SaratogaHouses.csv')
#print(display(datos.head()))
result = filter(lambda x: x == 'Yes', datos['centralAir'])

result_list = list(result)

for value in result_list:
    print(value)
```
