<b> Iterando con .iterrows() </b>

- Devuelve cada fila del Dataframe como una tuple de pares (indice, serie pandas )

```python
import pandas as pd 
import numpy as np 
```

```python
pit_df = pd.read_csv('https://assets.datacamp.com/production/repositories/3581/datasets/779033fb8fb5021aee9ff46253980abcbc5851f3/baseball_stats.csv')
```

```python
print(display(pit_df))
```

<b> Utilizando .iterrows() </b>

```python
for i, row in pit_df.iterrows():
    print(row)
    print('\n')
```

```python
for i, row in pit_df.iterrows():
    print(i)
    print(row)
    print(type(row))
    print('\n')
```

```python
for row_tuple in pit_df.iterrows():
    print(row_tuple)
    print(type(row_tuple))
```

<b> Conclusión : </b>

- .iterrows() devuelve cada fila del DataFrame como una tupla de pares (índice, Serie pandas), puedes dividir esta tupla y utilizar el índice y los valores de la fila por separado (como hiciste con for i,row en pit_df.iterrows()), o puedes mantener el resultado de .iterrows() en forma de tupla (como hiciste con for row_tuple en pit_df.iterrows()).

- Si utiliza i,row, puede acceder a cosas de la fila utilizando corchetes (es decir, row['Equipo']). Si se utiliza row_tuple, habría que especificar a qué elemento de la tupla se quiere acceder antes de coger el nombre del equipo (es decir, row_tuple[1]['Equipo']).

- Con cualquiera de los dos enfoques, el uso de .iterrows() seguirá siendo sustancialmente más rápido que el uso de .iloc

<b>.apply()  </b>

- Esta función actúa como la función de map() que hemos usado. Toma una función de entrada y la aplica a todo el DataFrame

- Para especificar en donde queremos aplicar .apply(), utilizamos 0 para las columnas y 1 para los renglones. con axis = 1 ó 0 

```python
#Ejercicio 

# Obtener la suma de tres columnas utilizando .apply()

stats_total = pit_df[['RS','RA','W']].apply(sum, axis = 0)
print('\n')
print(stats_total)


```

```python
#Ejercicio 

# Obtener la suma de dos renglones de RS Y RA  utilizando .apply()

stats_total = pit_df[['RS','RA']].apply(sum, axis = 1)
print('\n')
print(stats_total)


```

<b> Utilizando funciones lambda. </b>

```python
dict = {     
        2012: {'RS' :697, 'RA':  577, 'W' : 90,  'Playoffs' : 0},
        2011: {'RS' :707, 'RA':  614, 'W' : 91,  'Playoffs' : 1},
        2010 : {'RS' : 802, 'RA':  649, 'W': 96,  'Playoffs' : 1},
        2009 :{'RS' : 803, 'RA': 754, 'W':  84,  'Playoffs' : 0},
        2008 :{'RS' :774, 'RA': 671, 'W':  97,   'Playoffs' : 1}}
```

```python
dict_df = pd.DataFrame(dict)

df = dict_df.transpose()
df
```

```python
# Gather sum of all columns
stat_totals = df.apply(sum, axis=0)
print(stat_totals)
```

```python
# Gather total runs scored in all games per year
total_runs_scored = df[['RS', 'RA']].apply(sum, axis=1)
print(total_runs_scored)
```

```python
def text_playoffs(num_playoffs):
    if num_playoffs == 1:
        return 'yes'
    else:
        return 'no'
```

```python
# Convert numeric playoffs to text by applying text_playoffs()
textual_playoffs = df.apply(lambda row: text_playoffs(row['Playoffs']), axis=1)
print(textual_playoffs)
```

<b> Conclusión : </b>

- El método .apply() permite aplicar funciones a todas las filas o columnas de un DataFrame especificando un eje.

- Si has estado usando pandas durante algún tiempo, habrás notado que una mejor manera de encontrar estas estadísticas sería usar el método incorporado de pandas .sum().

<b> Resuelve un debate con .apply() </b>

- Los Diamondbacks de Arizona se han enterado de tus increíbles habilidades analíticas. 
  Les gustaría que les ayudaras a resolver un debate entre los directivos. Un entrenador afirma que el equipo ha llegado a los playoffs todos los años en los que ha tenido un porcentaje de victorias de 0,50 o superior. Otro mánager dice que esto no es cierto.


```python
# Primero obtengamos el dataframe exclusivamente del equipo 'ARI'
ari_team = pit_df[pit_df['Team'] == 'ARI'] 
dr = ari_team.iloc[:, 0:13]
```

```python
# Eliminamos tablas que no necesitamos
dr.drop(['SLG','BA', 'RankSeason'	,'RankPlayoffs', 'OBP'], axis = 1)
```

```python
# Reordenamos las columnas
d_df = dr[['Team',	'League',	'Year',	'RS',	'RA',	'W',	'G', 'Playoffs']]
```

```python
# Creamos una nueva columna que sera nuestro  nuevo indice 
d_df['index'] = range(15)
d_df.set_index(['index'], inplace = True)
# Ya tenemos nuestro DataFrame para trabajar 
d_df
```

 <b>Utilicemos la siguiente función y el método .apply() para ver qué mánager tiene razón. </b>

```python
def calc_win_perc(wins, games_played):
    win_perc = wins/ games_played
    return np.round(win_perc,2)
```

> Crea una Serie pandas llamada win_percs aplicando la función calc_win_perc() a cada fila del DataFrame con una función lambda.

```python
win_percs = d_df.apply(lambda row: calc_win_perc(row['W'], row['G']), axis = 1)
print(win_percs, '\n')
```

```python
# Agregando una nueva columna con los nuevos calculos
d_df['WP'] = win_percs
d_df
```

```python
# Filtrando los equipos donde el WP  sea mayor que .50

d_final = d_df[d_df['WP'] > .50]
d_final
```

<b> Conclusiones: </b>
- EL que tiene razón es el gerente que afirmó que el equipo no ha llegado a los playoffs todos los años que ha tenido un porcentaje de victorias de 0,50 o más.

- Lamentablemente, el segundo gestor era correcto. En los años 2012, 2008, 2003 y 2000 los Diamondbacks de Arizona tuvieron un porcentaje de victorias mayor o igual a 0,50, pero aun así no llegaron a los playoffs.


- El uso del método .apply() con una función lambda permite aplicar una función a un DataFrame sin necesidad de escribir un bucle for.


### Recordemos que Pandas esta basado en numpy 

- Teniendo esto en cuenta podemos volvernos mas eficacez en nuestros calculos porque podemos vectorizar nuestros calculos en vez de ir fila por fila, podemos aplicar nuestros calculos a todos los renglones en un solo movimiento. 

<b> Utilizando .values podemos vectorizar todos los datos, haciendo que estos se vuelvan un numpy array </b>

```python
# El problema anterior podemos resolverlo con mayor rapidez haciendo lo siguiente
win_perc_np = calc_win_perc(d_df['W'].values, d_df['G'].values)


d_df['WP_2'] = win_perc_np
d_df
```

- El uso de las matrices subyacentes de un DataFrame para realizar cálculos puede realmente acelerar su código y produce algunas ganancias significativas de eficiencia.
