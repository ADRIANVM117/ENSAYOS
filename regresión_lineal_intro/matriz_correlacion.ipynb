# Matriz de correlaci√≥n

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
testing = pd.read_csv('https://raw.githubusercontent.com/adolfollin/Bourbaki/master/primerCurso/Datos/datos-testingSet.csv')
training = pd.read_csv('https://raw.githubusercontent.com/adolfollin/Bourbaki/master/primerCurso/Datos/datos-trainingSet.csv')
```

```python
training.corr()
```

## Visualizaci√≥n mapa de calor 

```python
correlacion = training.corr()
correlacion = correlacion.round(2)
```

```python
f, ax = plt.subplots(figsize = (18,18))
temp = sns.heatmap(correlacion, mask = np.zeros_like(correlacion, dtype = np.bool), 
                   cmap = sns.diverging_palette(220, 10, as_cmap = True),
                   square = True, ax = ax, annot = True) 
```

## Feature enginering 
- Obtener nuevas variables apartir de las ya existentes

```python
training['Years'] = training.YrSold - training.YearBuilt
```

## Analisis de las mejores variables

```python
variables_continuas = correlacion.SalePrice[correlacion.SalePrice > .4].index.tolist()[:-1] + ['Years']
```

```python
correlacion = training.corr().loc[variables_continuas, variables_continuas]
```

```python
f, ax = plt.subplots(figsize = (16,16))
temp = sns.heatmap(correlacion, mask = np.zeros_like(correlacion, dtype = np.bool), 
                   cmap = sns.diverging_palette(220, 10, as_cmap = True),
                   square = True, ax = ax, annot = True) 
```

```python
variables_continuas = list(set(variables_continuas) - set(['YearBuilt', 'GarageYrBlt', '1stFlrSF', 
                                                           'TotRmsAbvGrd', 'GarageCars']))
```
